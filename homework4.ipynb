W4A20 Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không 

```python
n = int(input())
if n < 1:
    print('Không có dạng 2 ** k')
else:
    while n % 2 == 0 and n > 1:
        n //= 2
    if n == 1:
        print('True')
    else:
        print('False')
```

W4A21 Nhập vào số nguyên dương n, tính tổng các chữ số của n 

```python
n = int(input())
sum = 0
while n > 0:
    chu_so_cuoi = n % 10
    sum += chu_so_cuoi
    n //= 10
print(sum)
```

W4A22 Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ 

```python
n = int(input())
chan = 0
le = 0
while n > 0:
    chu_so = n % 10
    if chu_so % 2 == 0:
        chan += 1
    else:
        le += 1
    n //= 10
print(f'Số chữ số chẵn là : {chan}')
print(f'Số chữ số lẻ là : {le}')
```

Nhập vào A, tìm n nhỏ nhất sao cho 
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A 

```python
A = int(input())
sum = 0
n = 0
while sum < A:    #Tổng còn nhỏ hơn thì tiếp tục cộng
    n += 1 
    sum += 1/n
print(n)
```

W4A25 
Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1. 
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập

```python
max_val = None
min_val = None
while True:
    n = int(input())
    if n == -1:
        break
    if max_val is None:
        max_val = n
        min_val = n
    else:
        if n > max_val:
            max_val = n
        if n < min_val:
            min_val = n
if max_val is not None:
    print('Số lớn nhất là : ', max_val)
    print('Số nhỏ nhất là : ', min_val)
else:
    print('Bạn chưa nhập số nào hợp lệ')
        
```

W4A26 
[Fibonaci1] 
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ 
bằng tổng hai số trước đó 
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt 
quá A 

```python
A = int(input())
a = 0
b = 1
so_lon_nhat = 0
if A < 0:
    print('False')
elif A == 1:
    print('Số Fibonacci lớn nhất là 0')
else:
    while b < A:
        so_lon_nhat = b
        tong = a + b
        a = b
        b = tong
    print(f'Số Fibonacci lớn nhất nhưng không vượt quá {A} là : {so_lon_nhat}')
        
```

W4A27 
Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu từ (quy định là chuỗi không có 
ký tự đặc biệt, không số, không có dấu câu, chỉ có ký tự chữ và khoảng trắng) 
Dùng cách: for trên index; KHÔNG dùng hàm split(), len() 

```python
a = str(input())
count = 0
length = 0
dang_trong_tu = False
for ki_tu in a:   #Tìm độ dài chuỗi a
    length += 1
for i in range(length):
    if a[i] != ' ' and not dang_trong_tu:
        count += 1
        dang_trong_tu = True
    elif a[i] == ' ':
        dang_trong_tu = False
print('Số từ của chuỗi là : ', count)
```

W4A29 
Cho một chuỗi chỉ gồm các ký tự chữ cái thường và chữ cái hoa, hãy đổi ký tự hoa thành 
thường và thường thành hoa.

```python
a = str(input())
ket_qua = ' '
for kitu in a:
    if kitu.isupper():
        kitu = kitu.lower()
    elif kitu.islower():
        kitu = kitu.upper()
    ket_qua += kitu
print(ket_qua)
```

W4A31 
Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in 
thường, bao nhiêu ký tự số

```python
a = str(input())
hoa = 0
thuong = 0
so = 0
length = 0
for ki_tu in a:
    length += 1
    if ki_tu.isupper():
        hoa += 1
    elif ki_tu.islower():
        thuong += 1
    elif ki_tu.isdigit():
        so += 1
print(f'Số kí tự hoa, thường, số lần lượt là : {hoa}, {thuong}, {so}')
```
