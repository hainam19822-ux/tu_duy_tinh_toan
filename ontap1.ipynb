1 Yêu cầu. Cước phí Taxi Thông Minh. 
Viết một hàm duy nhất smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính và trả về total (tổng tiền phải trả, 
kiểu số nguyên, đơn vị VND). 
 
Các quy tắc cần cài đặt: 
1. Giá mở cửa (0–0.8 km): - Xe 4 chỗ: 11,000 VND - Xe 7 chỗ: 12,000 VND 
2. Giá theo km sau 0.8 km: - Xe 4 chỗ: từ 0.8→20 km nh 13,000 VND/km; trên 20 km tính 10,500 VND/km - Xe 7 chỗ: từ 0.8→20 km nh 14,000 VND/km; trên 20 km tính 12,000 VND/km 
3. Phí chờ: 5 phút đầu miễn phí; sau đó, mỗi 5 phút (làm tròn lên theo từng block 5 phút) tính 2,000 VND. 
4. Phụ thu ban đêm: nếu start_hour thuộc [22,23,0,1,2,3,4,5], cộng +20% trên tiền tạm tính (mở cửa + quãng đường + 
chờ). 
5. Điều kiện đặc biệt: - Nếu km > 30 và không phải ban đêm, áp dụng giảm 10% trên tổng tiền. - Nếu tổng tiền cuối cùng < 50,000 nhưng km > 5, đặt tổng tiền bằng 50,000 VND (mức tối thiểu).

```python
import math
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    if car_type == 4:
        gia_mo_cua = 11000
        gia_den_km_20 = 13000
        gia_tren_km_20 = 10500
    elif car_type == 7:
        gia_mo_cua = 12000
        gia_den_km_20 = 14000
        gia_tren_km_20 = 12000
    else:
        return 'Không có loại xe hợp lệ'
    
    #Tính tiền theo số km đi được
    so_tien_theo_km = 0
    if 0 <= km <= 0.8:
        so_tien_theo_km = gia_mo_cua
    elif 0.8 < km <= 20:
        so_tien_theo_km = gia_mo_cua + (km-0.8) * gia_den_km_20
    else:
        so_tien_theo_km = gia_mo_cua + (20-0.8) * gia_den_km_20 + (km - 20) * gia_tren_km_20
    
    #Tính phí chờ
    phi_cho = 0
    if wait_minutes <= 5:
        phi_cho = 0
    if wait_minutes > 5:
        thoi_gian_them = wait_minutes - 5
        block = math.ceil(thoi_gian_them / 5)
        phi_cho = block * 2000
    
    #Tổng tiền chưa tính phụ thu ban đêm 
    total = so_tien_theo_km + phi_cho
        
    #Tính phí phụ thu ban đêm
    gio_ban_dem = [22, 23, 0, 1, 2, 3, 4, 5]
    if start_hour in gio_ban_dem:
        total *= 1.2
    
    #Tính điều kiện đặc biệt
    if km > 30 and start_hour not in gio_ban_dem:
        total *= 0.9
    if total < 50000 and km > 5:
        total = 50000
        
    return int(total)

car_type, start_hour, km, wait_minutes = map(int, input().split())
tong_tien = smart_taxi_fare(car_type, start_hour, km, wait_minutes)
print(f'Tổng số tiền phải trả khi chọn xe {car_type} chỗ đi lúc {start_hour} giờ, đi {km} km và chờ {wait_minutes} phút là : {tong_tien}')      
```

4 Yêu cầu: Cho các hệ số thực a, b, c, trả về tất cả các nghiệm thực của phương trình bậc hai axx + bx + c = 0 trong một danh sách được sắp xếp tăng dần.
Quy tắc: 
- Nếu a == 0 và b != 0: trả về nghiệm duy nhất [-c / b]. 
- Nếu a == 0 và b == 0: trả về [] (không có nghiệm hữu hạn duy nhất nếu c == 0, hoặc vô nghiệm nếu c != 0). 
- Với a != 0, dùng biệt thức d = bb - 4ac:
 - d < 0 → [] 
 - d == 0 → [ -b / (2a) ] 
 - d > 0 → hai nghiệm [r1, r2] với r1 <= r2.

```python
def giai_phuong_trinh_bac_hai(a, b, c):
    if a == 0 and b != 0:
        return f'[{-c/b}]'
    elif a == 0 and b == 0 and c == 0:
        return '[]' 
    elif a == 0 and b == 0 and c != 0:
        return '[]' 
    else:
        denta = b**2 - 4*a*c
        if denta < 0:
            return '[]'
        elif denta == 0:
            return f'[{-b/(2*a)}]' 
        else:
            x1 = (-b + denta**0.5) / (2*a)
            x2 = (-b - denta**0.5) / (2*a)
            if x1 < x2:
                return f'[{x1}, {x2}]'
            else:
                return f'[{x2}, {x1}]'

a, b, c = map(float, input().split())
print(giai_phuong_trinh_bac_hai(a, b, c))
```

16 Viết hàm trả về góc ∠ABC theo đơn vị độ khi biết tọa độ các điểm A(x1,y1), B(x2,y2), C(x3,y3).

```python
import math
def tinh_goc(x1, y1, x2, y2, x3, y3):
    a = ((x3-x2)**2 + (y3-y2)**2) ** 0.5  #BC
    b = ((x3-x1)**2 + (y3-y1)**2) ** 0.5  #AC
    c = ((x2-x1)**2 + (y2-y1)**2) ** 0.5  #AB
    cos_ABC = (a**2 + b**2 - c**2) / (2*a*b)
    goc_ABC = math.degrees(math.acos(cos_ABC))
    return goc_ABC
x1, y1, x2, y2, x3, y3 = map(float, input().split())
goc_ABC = tinh_goc(x1, y1, x2, y2, x3, y3)
print(f'Số đo góc ABC là : {goc_ABC:.1f} độ')
    
```

- Đầu vào: - quad: danh sách hoặc bộ gồm đúng 4 đỉnh [(x1,y1), (x2,y2), (x3,y3), (x4,y4)] theo thứ tự (tứ giác có thể lồi hoặc lõm, nhưng không tự cắt). 
- p: một điểm (x, y). 
- Đầu ra: giá trị boolean. 
- Gợi ý: Cài đặt phép kiểm tra điểm trong đa giác tổng quát (ví dụ: ray casting), đồng thời tính cả trường hợp điểm nằm 
trên biên là ở trong. 

```python
def point_in_quadrilateral(x1, y1, x2, y2, x3, y3, x4, y4, x, y):
    def area(x1, y1, x2, y2, x3, y3):
        return abs((x1*(y3-y2) + x2*(y3-y1) + x3*(y2-y1)) / 2)
    A = area(x1, y1, x2, y2, x4, y4) + area(x2, y2, x3, y3, x4, y4)
    A1 = area(x, y, x1, y1, x2, y2)
    A2 = area(x, y, x1, y1, x4, y4)
    A3 = area(x, y, x3, y3, x4, y4)
    A4 = area(x, y, x2, y2, x3, y3)
    return A >= A1 + A2 + A3 + A4
x1, y1, x2, y2, x3, y3, x4, y4, x, y = map(float, input().split())
if point_in_quadrilateral(x1, y1, x2, y2, x3, y3, x4, y4, x, y):
    print('True')
else:
    print('False')
```

Yêu cầu: Viết hàm quadratic_roots_count(a, b, c) trả về số nghiệm thực của phương trình ax^2 + bx + c = 0. 
- Đầu vào: các số thực a, b, c.
 - Đầu ra: 0, 1, hoặc 2. 
 Nếu phương trình suy biến thành 0 = 0 (tức a == 0, b == 0, c == 0), trả về float('inf').
  - Quy tắc: 
  - Nếu a != 0, dùng biệt thức D = b^2 - 4ac. 
  - Nếu a == 0 và b != 0, đó là phương trình bậc nhất với đúng một nghiệm thực.
   - Nếu a == 0 và b == 0 và c != 0, phương trình vô nghiệm. 

```python
def quadratic_roots_count(a, b, c):
    if a == 0 and b == 0 and c == 0:
        return "float('inf')"
    elif a == 0 and b == 0 and c != 0:
        return '0' 
    elif a == 0 and b != 0 and c != 0:
        return '1'
    else:
        denta = b ** 2 - 4 * a * c
        if denta < 0:
            return '0' 
        elif denta == 0:
            return '1' 
        else:
            return '2' 
a, b, c = map(float, input().split())
print(quadratic_roots_count(a, b, c))
```

```python
def sum_halves(s):
    length = len(s)
    if length % 2 == 0:
        ba_chu_so_dau = int(s[:length // 2])
        ba_chu_so_cuoi = int(s[length // 2:])
        return ba_chu_so_dau + ba_chu_so_cuoi
    return 'Nhập chuỗi có độ dài chẵn' 
s = input()
print(f'Tổng của ba chữ số đầu và ba chữ số cuối của {s} là : {sum_halves(s)}')
```

chấm động, ngược lại trả về 'neither'. 
- Đầu vào: chuỗi s. - Đầu ra: một trong các giá trị {'int', 'float', 'neither'}. 
- Quy tắc: - Cho phép có dấu + hoặc - ở đầu. 
- Số nguyên phải chỉ gồm các chữ số và không có dấu chấm thập phân (ví dụ: '100'). 
- Số thực phải có dấu chấm và có chữ số ở cả hai phía của dấu chấm (ví dụ: '100.0', '-3.14'). 
- Các chuỗi như '.5', '5.', '1e3' được xem là 'neither' trong bài

```python
def detect_number_type(s):
    try:
        int(s)
        return "int" 
    except ValueError:
        try:
            float(s)
            return "float" 
        except ValueError:
            return "neither" 

s = input()
result = detect_number_type(s)
print(result)
```

32 Yêu cầu: Viết hàm is_right_triangle(a, b, c) trả về True nếu a, b, c có thể là ba cạnh của một tam giác vuông, ngược lại trả 
về False. 
- Đầu vào: các số dương a, b, c. 
- Đầu ra: giá trị boolean. 
- Lưu ý: Kiểm tra bất đẳng thức tam giác và định lý Pythagore. Dùng sai số nhỏ khi so sánh số thực. 

```python
import math
def is_right_triangle(a, b, c):
    if a <= 0 or b <= 0 or c <= 0:
        return False
    sides = sorted([a, b, c])   #Sắp xếp các cạnh để tìm cạnh lớn nhất
    a, b, c = sides[0], sides[1], sides[2]
    if a + b <= c:
        return False
    return math.isclose(a**2 + b**2, c**2, rel_tol = 1e-9)   #So sánh a**2 + b**2 có xấp xỉ c**2 hay không, rel_tol = 1e-9 2 số lệch nhau k quá 10^-9
a, b, c = map(float, input().split())
print(is_right_triangle(a, b, c))
```

33 Yêu cầu: Viết hàm compute(a, b, op) tính a (+|-||/) b tùy theo op.

- Nếu op là '/' và b == 0, phát sinh ValueError. 
- Nếu op không hợp lệ, phát sinh ValueError. 

```python
def compute(a, b, op):
    if op == '+':
        return a + b
    elif op == '-':
        return a - b
    elif op == '*':
        return a * b
    elif op == '/':
        if b == 0:
            return ValueError
        else:
            return a / b
    else:
        return ValueError
a, b = map(float, input().split())
op = input()
print(compute(a, b, op))
```

35 Yêu cầu: Viết hàm divide_with_remainder(dividend, divisor) trả về bộ (q, r) trong đó q là thương nguyên và r là số dư sao 
cho dividend = divisorq + r với 0 <= |r| < |divisor| (dùng ngữ nghĩa của Python với // và %). 
- Đầu vào: số nguyên dividend, divisor (divisor khác 0). 
- Đầu ra: bộ hai số nguyên (q, r). - Ví dụ: divide_with_remainder(5, 3) -> (1, 2).

```python
def divide_with_remainder(dividend, divisor):
    if divisor == 0:
        return False
    else:
        q = dividend // divisor
        r = dividend % divisor
        if divisor * q + r == dividend and abs(r) < abs(divisor):
            return q, r
dividend, divisor = map(int, input().split())
print(divide_with_remainder(dividend, divisor))
```

tạo thành tam giác. 

```python
def dien_tich_tam_giac(a, b, c):
    if a <= 0 or b <= 0 or c <= 0:
        raise ValueError('Các cạnh tam giác phải lớn hơn 0')
    p = (a+b+c) / 2
    return (p * (p-a) * (p-b) * (p-c)) ** 0.5
a, b, c = map(float, input().split())
dien_tich = dien_tich_tam_giac(a, b, c)
print(f'Diện tích tam giác ứng với 3 cạnh {a}, {b}, {c} là : {dien_tich:.1f}')
```

```python
def so_luong_k(n):
    dem = 0
    if n < 1:
        return False
    for i in range(1, n+1):
        if i < n and i > (n-1) / 2:
            dem += 1
    return dem
n = int(input())
print(f'Số lượng số nguyên dương thỏa mãn nếu n = {n} là : {so_luong_k(n)}')
```
