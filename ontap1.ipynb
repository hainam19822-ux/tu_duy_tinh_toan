2 Yêu cầu: Cho số nguyên x, trả về True nếu tồn tại số nguyên n ≥ 1 sao cho x = 1 + 2 + ... + n, ngược lại trả về False. 
Khuyến khích dùng vòng lặp đơn giản để cộng dồn tổng. 

```python
x = int(input("Nhập một số nguyên: "))
n = 1
tong = 0
while tong < x:
    n += 1
    tong += n
    
if tong == x:
    print("True")
else:
    print("False")
```

3 Yêu cầu: Cho tọa độ ba điểm A(x1,y1), B(x2,y2), C(x3,y3), trả về True nếu chúng tạo thành một tam giác, ngược lại trả về False. 
Quy tắc: Chúng tạo thành tam giác nếu không thẳng hàng (diện tích có hướng khác 0). 

```python
def is_triangle(x1, y1, x2, y2, x3, y3):
    dien_tich = 0.5 * abs(x1*(y2-y3) + x2*(y3-y1) + x3*(y2-y1))
    if dien_tich > 0:
        return True
    else:
        return False
x1, y1, x2, y2, x3, y3 = map(float, input().split())
print(is_triangle(x1, y1, x2, y2, x3, y3))
```

4 Yêu cầu: Cho các hệ số thực a, b, c, trả về tất cả các nghiệm thực của phương trình bậc hai axx + bx + c = 0 trong một danh sách được sắp xếp tăng dần.
Quy tắc: 
- Nếu a == 0 và b != 0: trả về nghiệm duy nhất [-c / b]. 
- Nếu a == 0 và b == 0: trả về [] (không có nghiệm hữu hạn duy nhất nếu c == 0, hoặc vô nghiệm nếu c != 0). 
- Với a != 0, dùng biệt thức d = bb - 4ac:
 - d < 0 → [] 
 - d == 0 → [ -b / (2a) ] 
 - d > 0 → hai nghiệm [r1, r2] với r1 <= r2.

```python
def giai_phuong_trinh_bac_hai(a, b, c):
    if a == 0 and b != 0:
        return f'[{-c/b}]'
    elif a == 0 and b == 0 and c == 0:
        return '[]' 
    elif a == 0 and b == 0 and c != 0:
        return '[]' 
    else:
        denta = b**2 - 4*a*c
        if denta < 0:
            return '[]'
        elif denta == 0:
            return f'[{-b/(2*a)}]' 
        else:
            x1 = (-b + denta**0.5) / (2*a)
            x2 = (-b - denta**0.5) / (2*a)
            if x1 < x2:
                return f'[{x1}, {x2}]'
            else:
                return f'[{x2}, {x1}]'

a, b, c = map(float, input().split())
print(giai_phuong_trinh_bac_hai(a, b, c))
```

5 Yêu cầu: Cho ba số a, b, c, trả về chúng dưới dạng danh sách theo thứ tự không giảm. 

```python
def thu_tu_khong_giam(a, b, c):
    if a >= b and a >= c:
        if b >= c:
            return f'[{a}, {b}, {c}]'
        else:
            return f'[{a}, {c}, {b}]' 
    elif b >= a and b >= c:
        if a >= c:
            return f'[{b}, {a}, {c}]'
        else:
            return f'[{b}, {c}, {a}]'
    else:
        if a >= b:
            return f'[{c}, {a}, {b}]'
        else:
            return f'[{c}, {b}, {a}]'
a, b, c = map(int, input().split())
print(thu_tu_khong_giam(a, b, c))
```

6 Yêu cầu: Cho năm số a, b, c, d, e, trả về True nếu tồn tại bất kỳ ba số nào trong đó bằng nhau; ngược lại trả về False. 
Không dùng tuple hay dictionary. 

```python
def ba_so_bang_nhau(a, b, c, d, e):
    danh_sach = [a, b, c, d, e]
    for so in danh_sach:
        if danh_sach.count(so) >= 3:   #.count : đếm số lần xuất hiện của số trong danh sách
            return True
        else:
            return False
a, b, c, d, e = map(int, input().split())
print(ba_so_bang_nhau(a, b, c, d, e))
```

7 Yêu cầu: Cho ba số a, b, c, trả về giá trị ở giữa khi sắp xếp ba số theo thứ tự tăng dần (giá trị lớn thứ hai). 
Định nghĩa này vẫn đúng khi có phần tử trùng nhau (ví dụ: 2, 2, 5 → giá trị giữa là 2).

```python
def so_o_giua(a, b, c):
    if a <= b and a <= c:
        if b >= c:
            return f'[{c}]'
        else:
            return f'[{b}]'
    elif b <= a and b <= c:
        if a >= c:
            return f'[{c}]'
        else:
            return f'[{a}]' 
    else:
        if a >= b:
            return f'[{b}]'
        else:
            return f'[{a}]'
a, b, c = map(int, input().split())
print(so_o_giua(a, b, c))
```
