Viết hàm kiểm tra một số có phải số Amstrong hay không? Hàm trả về kiểu bool, có tham số là một số 
nguyên x 
#Một số nguyên dương có n chữ số được gọi là số Armstrong nếu tổng các chữ số của nó, mỗi số mũ n lên, bằng chính nó.Ví dụ với số có 3 chữ số (n=3):Số 153 có 3 chữ số.Ta tính: 1^3 + 5^3 + 3^3 = 1 + 125 + 27 = 153.=> 153 là số Armstrong.

```python
def is_Amstrong_number(num):
    num_str = str(num)   #chuyển số thành chuỗi
    num_digits = len(num_str)  #đếm số chữ số của num
    tong = 0   #khởi tạo biến tổng
    for i in range(num_digits):  #lặp qua từng chữ số của num
        tong += int(num_str[i]) ** num_digits   #tính tổng các chữ số được nâng lên lũy thừa số chữ số
    return tong == num   #kiểm tra tổng có bằng num hay k
if __name__ == "__main__":   
    num = int(input())
    if is_Amstrong_number(num):
        print(f'{num} là số Amstrong')
    else:
        print(f'{num} không phải là số Amstrong')
    
    
```

Viết hàm kiêểm tra một số có phải số chính phương hay không? Hàm trả về kiểu bool, có tham số là một số nguyên x

```python
def is_square_number(n):
    if n < 0:
        return False
    else:
        square_root = int(n ** 0.5)
        return square_root * square_root == n
    
n = int(input())
if is_square_number(n):
    print(f'{n} là số chính phương')
else:
    print(f'{n} không phải số chính phương')
    
```

W5A18 
VIết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên

```python
def is_palindrome_number(n):
    if n < 0:
        return False
    orginal_n = n
    reverse_n = 0
    while n > 0:
        chu_so = n % 10
        reverse_n = reverse_n * 10 + chu_so
        n //= 10
    return orginal_n == reverse_n

n = int(input())
if is_palindrome_number(n):
    print(f'{n} là số đối xứng')
else:
    print(f'{n} không phải số đối xứng')
```

```python
def is_reverse_number(n):
    n_str = str(n)
    reverse_str = n_str[::-1]
    return n_str == reverse_str
n = int(input())
if is_reverse_number(n):
    print(f'{n} là số đối xứng')
else:
    print(f'{n} không phải là số đối xứng')
```

#giải thích dòng reverse_str = n_str[::-1]
đây là kĩ thuật cắt chuỗi (Slicing)
một lệnh đầy đủ có cấu trúc : sequenque[start:stop:step]
Dấu : thứ nhất và thứ hai (để trống start và stop): Nghĩa là chúng ta lấy toàn bộ chuỗi từ đầu đến cuối.
Tham số step là -1: Đây chính là "chìa khóa".
Thông thường step mặc định là 1 (đi từ trái sang phải).
Khi step là -1, Python sẽ đi ngược từ phải sang trái.

VIết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên. 

```python
def is_magic_number(n):
    while n >= 10:   #Nếu tổng còn 2 chữ số trở lên, tiếp tục tính tổng các chữ số
        sum = 0
        while n > 0:
            sum += n % 10
            n //= 10
        n = sum
    return n == 1

n = int(input())
if is_magic_number(n):
    print(f'{n} là số kì diệu')
else:
    print(f'{n} không phải là số kì diệu')
```

Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'. 

```python
def detect_number_type(s):
    try:
        int(s)
        return "int" 
    except ValueError:
        try:
            float(s)
            return "float" 
        except ValueError:
            return "neither" 

s = input()
result = detect_number_type(s)
print(result)

    
```

Viết hàm sum_halves(s) nhận chuỗi 6 chữ số và trả về tổng của hai số 3 chữ số đầu và 3 chữ số cuối.

```python
def sum_halves(s):
    length = len(s)
    if length % 2 == 0:
        ba_chu_so_dau = int(s[:length // 2])
        ba_chu_so_cuoi = int(s[length // 2:])
        return ba_chu_so_dau + ba_chu_so_cuoi
    else:
        return "Vui lòng nhập chuỗi có độ dài chẵn"
    
s = input()
result = sum_halves(s)
if len(s) % 2 == 0:
    print(result)
else:
    print("Vui lòng nhập chuỗi có độ dài chẵn")
```

W5A23
 Viết hàm quadratic_roots_count(a, b, c) trả về số nghiệm thực của phương trình a*x^2 + b*x + c = 0. 

```python
def quadratic_roots_count(a, b, c):
    denta = b**2 - 4*a*c
    if denta > 0:
        return "Số nghiệm thực phân biệt : 2"
    elif denta == 0:
        return "Số nghiệm thực bằng nhau : 1"
    else:
        return "Không có nghiệm thực"

if __name__ == "__main__":
    a, b, c = map(float, input().split())
    result = quadratic_roots_count(a, b, c)
    print(result)
```

Viết hàm point_in_quadrilateral(quad, p) kiểm tra điểm p có nằm trong hoặc trên biên của tứ giác quad hay không. Trong đó: quad là một tứ giác (4 đỉnh – 4 điểm), p là một điểm cần kiểm tra. VD: quad = [(0,0), (4,0), (4,3), 
(0,3)], p = (2,1) 
Nếu điểm P nằm trong hoặc trên biên tứ giác ABCD, thì tổng diện tích của 4 tam giác tạo bởi P với các cạnh tgPAB + tgPBC + tgPCD + tgPDA phải bằng diện tích của chính tứ giác ABCD.


```python
def point_in_quadrilateral(x1, y1, x2, y2, x3, y3, x4, y4, x, y):
    def area(x1, y1, x2, y2, x3, y3):
        return abs((x1*(y2-y3) + x2*(y3-y1) + x3*(y1-y2)) / 2.0)
    A = area(x1, y1, x2, y2, x4, y4) + area(x2, y2, x3, y3, x4, y4)
    A1 = area(x, y, x2, y2, x3, y3)
    A2 = area(x1, y1, x, y, x4, y4)
    A3 = area(x1, y1, x2, y2, x, y)
    A4 = area(x1, y1, x3, y3, x4, y4)
    return A == A1 + A2 + A3 + A4
if __name__ == "__main__":
    x1, y1, x2, y2, x3, y3, x4, y4, x, y = map(float, input().split())
    if point_in_quadrilateral(x1, y1, x2, y2, x3, y3, x4, y4, x, y):
        print("Điểm nằm trong hoặc trên cạnh của tứ giác")
    else:
        print("Điểm nằm ngoài tứ giác")
```

W5A27
 Viết hàm trả về số lượng chữ số khác 0 trong n số nguyên nhập vào từ bàn phím (một list)

```python
def dem_chu_so_khac_khong(n_count):
    chu_so_khac_khong = 0
    
    for i in range(n_count):
        n = int(input())
        n = abs(n)
        
        if n == 0:
            continue 
        temp = n
        while temp > 0:
            chu_so = temp % 10
            if chu_so != 0:
                chu_so_khac_khong += 1
            temp //= 10
    return chu_so_khac_khong

n = int(input())
print("Số chữ số khác 0 là :", dem_chu_so_khac_khong(n))
```

Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3). 

```python
def tinh_goc(x1, y1, x2, y2, x3, y3):
    import math
    a = math.sqrt((x2-x1)**2 + (y2-y1)**2)
    b = math.sqrt((x3-x2)**2 + (y3-y2)**2)
    c = math.sqrt((x3-x1)**2 + (y3-y1)**2)
    # Tính góc tại đỉnh B
    cos_ABC = (a**2 + b**2 - c**2) / (2 * a * b)
    goc_ABC = math.degrees(math.acos(cos_ABC))
    return goc_ABC
x1, y1, x2, y2, x3, y3 = map(float, input().split())
goc_ABC = tinh_goc(x1, y1, x2, y2, x3, y3)
print(f'Góc ABC có số đo là : {goc_ABC:.1f} độ')
```

Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau 
Nếu n ≤ 0 trả về 0

```python
def chu_so_doi_mot_khac_nhau(x):
    x_str = str(x)
    return len(set(x_str)) == len(x_str)
def sum_1_to_n(n):
    if n <= 0:
        return 0
    else:
        tong = 0
        for x in range(1, n+1):
            if chu_so_doi_mot_khac_nhau(x):
                tong += x
        return tong
n = int(input())
if n <= 0:
    print("Vui lòng nhập số dương")
else:
    print(f"Tổng các chữ số từ 1 đến {n} có các chữ số khác nhau là : {sum_1_to_n(n)}")
```
