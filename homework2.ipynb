Bài 1: Nhập vào 2 cạnh của một hình chữ nhật. In ra màn hình diện tích và chu 
vi của nó.  

```python
# chieu dai = a
# chieu rong = b
a, b = map(float, input('Nhập độ dài 2 cạnh :').split())
dien_tich = a * b
chu_vi = (a + b) * 2
print(f'Diện tích hình chữ nhật là : {dien_tich:.2f}')
print(f'Chu vi hình chữ nhật là : {chu_vi:.2f}')
```

Bài 2: Nhập vào bán kính của hình tròn. In ra màn hình diện tích và chu vi của 
nó. 

```python
# bán kính hình tròn = r
# pi = 3.14
pi = 3.14
r = float(input('Nhập bán kính :'))
print(f'Chu vi hình tròn là : {2 * pi * r:.2f}')
print(f'Diện tích hình tròn là : {pi * r * r:.2f}')
```

Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểm tra xem nó có thể là 
độ dài ba cạnh hay không, nếu  không thì in  ra màn  hình  'Khong la ba  canh 
cua mot tam  giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, 
Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra màn 
hình. 

```python
a, b, c = map(float, input('Nhập độ dài 3 cạnh :').split())
if a > 0 and b > 0 and c > 0 and a + b > c and a + c > b and b + c > a:
    print('Đây là 3 cạnh của tam giác')
    if a == b == c:
        print('Đây là tam giác đều')
    elif a ** 2 + b ** 2 == c ** 2:
        print('Đây là tam giác vuông')
    elif (a == b or b == c or a == c):
        print('Đây là tam giác cân')
    else:
        print('Đây là tam giác thường')
    chu_vi = a + b + c
    p = (a + b + c) / 2
    dien_tich = (p * (p - a) * (p - b) * (p - c)) ** 0.5
    print(f'Chu vi tam giác là : {chu_vi:.2f}')
    print(f'Diện tích tam giác là : {dien_tich:.2f}')
else:
    print('Đây không phải là tam giác')    
```

Bài 5: Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
a, b, c = map(float, input('Nhập 3 hệ số a b c : ').split())
if a != 0 and b != 0 and c != 0:
    denta = b ** 2 - 4 * a * c
    x1 = (-b + denta ** 0.5) / (2 * a)
    x2 = (-b - denta ** 0.5) / (2 * a)
    print(f'Phương trình có 2 nghiệm phân biệt là : x1 = {x1:.2f} và x2 = {x2:.2f}')
elif a == 0 and b != 0 and c != 0:
    x3 = -c / b
    print(f'Phương trình có nghiệm duy nhất là : {x3:.2f}')
elif a != 0 and b != 0 and c == 0:
    x4 = 0
    x5 = -b / a
    print(f'Phương trình có 2 nghiệm là : x4 = 0 và x5 = {x5:.2f}')
else:
    print('Phương trình vô nghiệm')
    
    
```

Bài 8: Giải hệ phương trình tuyến tính: 
ax + by = m 
cx + dy = n

```python
#D : định thức
a, b, m, c, d, n = map(float, input('Nhập các giá trị hệ số tương ứng :').split())
D = (a * d)-(b * c)
D1 = (m * d)-(b * n)
D2 = (a * n)-(m * c)
if D != 0 and D1 != 0 and D2 != 0:
    print(f'Hệ phương trình có nghiệm duy nhất là : x = {D1 / D:.1f} và y = {D2 / D:.1f}')
elif D == 0 and D1 != 0 and D2 != 0:
    print(f'Hệ phương trình có vô số nghiệm')
else:
    print(f'Hệ phương trình vô nghiệm')
    
    
```

Bài 10: Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vào một 
điểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
import math
r = float(input('Nhập bán kính r :'))
x, y = map(float, input('Nhập tọa độ x và y của điểm A :').split())
ban_kinh_binh_phuong = r ** 2
khoang_cach_binh_phuong = x ** 2 + y ** 2
print(f'Ta có : {x}**2 + {y}**2 == {khoang_cach_binh_phuong:.1f}')
print(f'Và {r}**2 == {ban_kinh_binh_phuong:.1f}')
if khoang_cach_binh_phuong <= ban_kinh_binh_phuong:
    print('Điểm A thuộc đường tròn')
else:
    print('Vì x**2 + y**2 > r**2 nên')
    print('Điểm A không thuộc đường tròn')

```
