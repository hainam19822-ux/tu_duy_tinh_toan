W5A3 Viết hàm kiểm tra một số có phải là số nguyên tố hay không?

```python
def check(n):
    if n < 2:
        return False
    else:
        for i in range(2, int(n ** 0.5) + 1):
            if n % i == 0:
                return False
        return True
        
n = int(input())
if check(n):
    print(f'{n} là số nguyên tố')
```

```python
def is_prime_number(n:int) -> bool:
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

assert is_prime_number(3) == True
assert is_prime_number(2) == True
assert is_prime_number(0) == False
assert is_prime_number(-10) == False
assert is_prime_number(17) == True

print('Pass all tests!')
```

W5A4 VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên.
# Số hoàn hảo là số có tổng các ước (trừ chính nó) bằng số đó

```python
def check(n):
    tong = 0
    for i in range(1, n):
        if n % i == 0:
            tong += i
    return tong == n
n = int(input())
if check(n):
    print(f'{n} là số hoàn hảo')
else:
    print(f'{n} là số không hoàn hảo')
```

```python
def is_perfect_number(n:int) -> bool:
    if n < 2:
        return False
    tong_uoc = 0
    for i in range(1, n):
        if n % i == 0:
            tong_uoc += i
    return tong_uoc == n
assert is_perfect_number(6) == True
assert is_perfect_number(9) == False
assert is_perfect_number(28) == True
assert is_perfect_number(496) == True
assert is_perfect_number(100) == False

print('Pass all tests!')
```

đầu tiên có giá trị k trong list số nguyên, nếu không có thì trả về -1

W5A6 Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n (n!)

```python
def giaithua(n):
    res = 1
    for i in range(1, n+1):
        res *= i
    return res

n = int(input())
print(giaithua(n))
```

Hàm thực hiện tính toán kết quả của phép toán giữa số \( num1 \) và \( num2 \) với toán tử \( operat

Biết các toán tử đơn giản thực hiện phép toán lần lượt là: 
Phép trừ \( - \) 
Phép cộng \( + \) 
Phép chia \( / \) 
Phép nhân \( * \) 
Kết quả làm tròn đến chữ số thập phân thứ hai. 
Đầu vào: Hai số thực và phép toán cần thực hiện (trong số các phép toán \(+, -, *, /\)). 
Đầu ra: Kết quả phép toán.

W5A8 
Khoảng cách Hamming giữa hai số nguyên là số lượng vị trí khác nhau giữa hai dãy bits tương ứng 
của chúng. Ví dụ: Khoảng cách Hamming giữa 1 (1) và 4 (100) là 2. 
Cho hai số nguyên  x và y, hãy viết hàm trả về khoảng cách Hamming giữa hai số này. 
Khoảng cách Hamming là cái tên được đặt theo tên của Richard Hamming, người giới thiệu lý 
thuyết này trong tài liệu có tính cơ sở của ông về mã phát hiện lỗi và sửa lỗi (error-detecting and 
error-correcting codes). Nó được sử dụng trong kỹ thuật viễn thông để tính số lượng các bit trong 
một từ nhị phân (binary word) bị đổi ngược, như một hình thức để ước tính số lỗi xảy ra trong quá 
trình truyền thông, và vì thế, đôi khi, nó còn được gọi là khoảng cách tín hiệu (signal distance). Việc 
phân tích trọng số Hamming của các bit còn được sử dụng trong một số ngành, bao gồm lý thuyết 
tin học, lý thuyết mã hóa, và mật mã học. Tuy vậy, khi so sánh các dãy ký tự có chiều dài khác 
nhau, hay các dãy ký tự có xu hướng không chỉ bị thay thế không thôi, mà còn bị ảnh hưởng bởi dữ 
liệu bị lồng thêm vào, hoặc bị xóa đi, phương pháp đo lường phức tạp hơn, như khoảng cách 
Levenshtein (Levenshtein distance) là một phương pháp có tác dụng và thích hợp hơn. 

```python
def ten_to_Bin(n:int) -> str:
    ans = ''
    while n > 0:
        du = n % 2
        ans = str(du) + ans
        n //= 2
    return ans

def Hamming(n:int, m:int) -> int:
    n_bin = ten_to_Bin(n)
    m_bin = ten_to_Bin(m)
    while len(n_bin) < len(m_bin):
        n_bin = '0' + n_bin
    while len(m_bin) < len(n_bin):
        m_bin = '0' + m_bin
    
    ans = 0
    for i in range(len(n_bin)):
        if m_bin[i] != n_bin[i]:
            ans += 1
    return ans

assert Hamming(1, 4) == 2
assert Hamming(123456, 32321) == 6
assert Hamming(87654321, 21321835) == 15
assert Hamming(8989, 8989) == 0

print('Pass all tests!')
```

```python
def khoang_cach_Hamming(x, y):
    if len(x) != len(y):
        return -1
    else:
        count = 0
        for i in range(len(x)):
            if x[i] != y[i]:
                count += 1
        return count
        

x = str(input())
y = str(input()) 
khoang_cach = khoang_cach_Hamming(x, y)
if khoang_cach != -1:
    print(f'Khoảng cách Hamming giữa {x} và {y} là: {khoang_cach}')
else:
    print('Chuỗi không cùng độ dài nên không thể tính khoảng cách Hamming')
```

Hai từ được gọi là đẳng cấu nếu các kí tự trong một từ có thể được ánh xạ để được từ còn lại. Ánh 
xạ một kí tự nghĩa là thay đổi toàn bộ các kí tự đó trong từ bởi một kí tự khác. Thứ tự của các kí tự 
được giữ không đổi. Hai kí tự khác nhau không thể ánh xạ đến cùng một kí tự, nhưng một kí tự có 
thể ánh xạ thành chính nó. Ví dụ, từ "abca" và "zbxz" là hai từ đẳng cấu vì ta có thể ánh xạ 'a' thành 
'z', 'b' thành 'b' và 'c' thành 'x'. 
Viết hàm  trả về \(true\) nếu hai từ \(a\) và \(b\) là đẳng cấu, ngược lại \(false\). Bạn có thể giả sử hai 
từ đầu vào có cùng độ dài. 

```python
def dang_cau(a, b):
    if len(a) != len(b):
        return False
    else:
        for i in range(len(a)):
            if a[i] != b[i] or a[i] == b[i]:
                return True
            
a = str(input())
b = str(input())
if dang_cau(a, b):
    print('True')
else:
    print('False')
```

W5A12 
Viết hàm divide_with_remainder(dividend, divisor) trả về bộ (q, r) sao cho dividend = divisor*q + r và 
0 <= |r| < |divisor|. 
# divindend : số bị chia
# divisor : số chia
# q : chia lấy phần nguyên 
# r : chia lấy phần dư
# => số bị chia = số chia * q + r

```python
def divide_with_remainder(divindend, divisor):
    if divisor == 0:
        return False
    else:
        q = divindend // divisor
        r = divindend % divisor
        return q, r
    
divindend = int(input("Nhập số bị chia: "))
divisor = int(input("Nhập số chia: "))
kqua = divide_with_remainder(divindend, divisor)
if kqua:
    q, r = kqua
    print(f'Kết quả phép chia {divindend} cho {divisor} là : {q} với số dư {r}')
else:
    print('Không thể thực hiện phép chia cho 0')
```

W5A14 Viết hàm compute(a, b, op) thực hiện phép tính a (+|-|*|/) b tùy theo ký tự op. 

```python
def compute(a, b, op):
    if op == '+':
        return a + b
    elif op == '-':
        return a - b
    elif op == '*':
        return a * b
    elif op == '/':
        if b != 0:
            return a/ b
        else:
            return False
        
a, b = map(float, input().split())
op = input()
ketqua = compute(a, b, op)
if ketqua is not False:
    print(f'Kết quả của phép tính {a} {op} {b} là : {ketqua:.1f}')
else:
    print('Không thể thực hiện phép chia cho 0')
```

W5A15 Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông. 

```python
def is_right_triangle(a, b, c):
    if a ** 2 + b ** 2 == c ** 2:
        return True
    else:
        return False
    
a, b, c = map(float, input().split())
if is_right_triangle(a, b, c):
    print(f'Ba số {a}, {b}, {c} là 3 cạnh của một tam giác vuông')
```
