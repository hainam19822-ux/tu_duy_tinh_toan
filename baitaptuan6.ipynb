sau khi xoay trái k vị trí. (Giữ nguyên thứ tự tương đối sau xoay; không sắp

W6A4 
Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo 
dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu key 
lặp lại, thêm vào list của key đó 

```python
key_value = input().split()
res_dict = {}
for i in key_value:
    key, value = i.split(':')
    if key not in res_dict:
        res_dict[key] = [value]
    else:
        res_dict[key].append(value)
print(res_dict)
```

W6A5 
Nhập một list số nguyên, In ra dictionary {'positives': ..., 'negatives': ..., 
'zeros': ...} (đếm số >0, <0, =0) 

```python
nums = list(map(int, input().split()))
res = {'positives':0, 'negatives':0, 'zeros':0}
for i in nums:
    if i > 0:
        res['positives'] += 1
    elif i < 0:
        res['negatives'] += 1
    else:
        res['zeros'] += 1
print(res)
```

Viết chương trình nhập vào một danh sách các số nguyên, và số nguyên 
dương k. In ra list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là 
một tuple, hai số trong tuple có thể giống nhau, thứ tự xuất hiện trong danh 
sách của số thứ nhất nhỏ hơn thứ tự của số thứ hai. Các tuple trong list được 
sắp xếp theo thứ tự tăng dần của phần tử đầu tiên trong tuple. 

```python
nums = list(map(int, input().split()))
k = int(input())
result = []
for i in range(len(nums)):
    for j in range(i+1, len(nums)):
        if nums[i] + nums[j] == k:
            cap_so = (nums[i], nums[j])
            if cap_so not in result:
                result.append(cap_so)
result.sort()
print(result)
```

Nhập vào một tuple chứa các số nguyên. Hãy tạo ra hai tuple mới: một chứa 
tất cả các số chẵn và một chứa tất cả các số lẻ từ tuple ban đầu, giữ nguyên 
thứ tự tương đối. In ra hai tuple này trên hai dòng riêng biệt. 

```python
tup = tuple(map(int, input().split()))
tup1 = []
tup2 = []
for i in tup:
    if i % 2 == 0:
        tup1.append(i)
    else:
        tup2.append(i)
print(tuple(tup1))
print(tuple(tup2))
```

```python
#1. Thành phần bên trong: (i for i in tup if i % 2 == 0)
Đây giống như một "bộ lọc" thông minh, nó có thể được hiểu theo ngôn ngữ tự nhiên là:
Lấy từng giá trị i nằm trong tup, nhưng chỉ lấy với điều kiện i chia hết cho 2.
Cấu trúc này gồm 3 phần:
i: Kết quả bạn muốn giữ lại.
for i in tup: Vòng lặp để duyệt qua toàn bộ dữ liệu ban đầu.
if i % 2 == 0: Bộ lọc điều kiện (chỉ cho phép số chẵn đi qua).

tup = tuple(map(int, input().split()))
tup_chan = tuple(i for i in tup if i % 2 == 0)
tup_le = tuple(i for i in tup if i % 2 != 0)
print(tup_chan)
print(tup_le)
```

list. Nếu có nhiều số cùng có số lần xuất hiện nhiều nhất, hãy in ra số nhỏ 
nhất trong số đó. 

```python
nums = list(map(int, input().split()))
count = {}
for i in nums:
    if i in count:
        count[i] += 1
    else:
        count[i] = 1
max_freq = max(count.values())
res = []
for key, lan_xuat_hien in count.items():
    if lan_xuat_hien == max_freq:
        res.append(key)
print(min(res))
```

[Đảo ngược key và value của dictionary] 
Nhập vào một dictionary có value là duy nhất (không trùng nhau). Hãy tạo 
một dictionary mới bằng cách đảo ngược key và value của dictionary ban 
đầu. In ra dictionary mới.

```python
k_v = input().split()
original_dict = {}
for i in k_v:
    k, v = i.split(':')
    original_dict[k] = v
reverse_dict = {}
for k, v in original_dict.items():
    reverse_dict[v] = k
print(reverse_dict)
```

[Lọc dictionary theo giá trị] 
Nhập vào một dictionary có key là chuỗi và value là số nguyên, và một số 
nguyên k ở dòng tiếp theo. Hãy tạo một dictionary mới chỉ chứa các cặp 
key-value từ dictionary ban đầu mà có value lớn hơn k. In ra dictionary 
mới.

```python
key_value = input().split()
k = int(input())
dic_ban_dau = {}
for i in key_value:
    key, value = i.split(':')
    dic_ban_dau[key] = value
dic_moi = {}
for key, value in dic_ban_dau.items():
    if int(value) > k:
        dic_moi[key] = value
print(dic_moi)
```
