Bài 9: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số chính phương hay 
không? Số chính phương là số có căn bậc hai là một số nguyên VD: 1, 4, 9, 16, 25…. 

```python
a = int(input('nhập số a :'))
is_square = False
for i in range (1, a+1):
    if i ** 2 == a:
        is_square = True
        break
if is_square:
    print(a, 'là số chính phương')
else:
    print(a, 'không là số chính phương')
```

 Nhập vào số nguyên dương a, kiểm tra xem a có phải là số hoàn hảo hay không.

Bài 13: Nhập vào số nguyên dương A, tìm n nhỏ nhất sao cho 
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A

```python
A = int(input())
sum_n = 0
n = 0
#Tổng chưa vượt quá A thì vẫn tiếp tục cộng
while sum_n < A:
    n += 1
    sum_n += 1/n
print(n)
```

```python
Bài 14: Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập 
xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng
```

```python
#Khởi tạo 1 giá trị ban đầu cho max và min
max_val = None
min_val = None
while True:
    n = int(input())
    if n == -1:
        break
    #Nếu chỉ nhập 1 số bất kì sau đó nhập -1 thì max = min
    if max_val is None:
        max_val = n
        min_val = n
    else:
        if n > max_val:
            max_val = n
        if n < min_val:
            min_val = n
#Kiểm tra có nhập số nào trc khi nhập -1 không
if max_val is not None:
    print('Số lớn nhất là :', max_val)
    print('Số nhỏ nhất là :', min_val)
else:
    print('Bạn chưa nhập số nào hợp lệ')
```

Bài 18: Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 3^k hay 
không 

```python
n = int(input())
if n < 1:
    print('False')
else:
    while n % 3 == 0 and n > 1:
        n //= 3
    if n == 1:
            print('True')
    else:
            print('False')
```

Bài 20: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố.

```python
n = int(input())
nguyen_to = 0
for i in str(n):
    chu_so = int(i)
    if chu_so == 2 or chu_so == 3 or chu_so == 5 or chu_so == 7:
        nguyen_to += 1
print('Số chữ số là số nguyên tố là : ', nguyen_to)
```

```python
n = int(input())
count = 0
while n > 0:
    chu_so = n % 10
    if chu_so > 1:
        is_prime = True
        for i in range (2,int(chu_so ** 0.5)+1):
            if chu_so % i == 0:
                is_prime = False
                break
        if is_prime:
            count += 1       
    n //= 10
print('Số chữ số nguyên tố là : ', count)
```

Bài 19: Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số 
kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong 
dãy số fibonacci lớn nhất nhưng không vượt quá A 

```python
A = int(input())
a = 0
b = 1
so_lon_nhat = 0
if A < 0:
    print('Không có số nào thỏa mãn')
elif A == 1:
    print('Số fibonacci lớn nhất là : 0')
else:
    while b < A:
        so_lon_nhat = b
        tong = a+b
        a = b
        b = tong
    print(f'Số lớn nhất trong dãy fibonacci nhưng không vượt quá {A} là : {so_lon_nhat}')
```
