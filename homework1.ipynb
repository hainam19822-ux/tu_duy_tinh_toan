[Rounding] 
Nhập vào một số thực: in ra kết quả làm tròn lên, tròn xuống, làm tròn tới số 
nguyên gần nhất (không được sử dụng hàm sẵn có của pyton). Các kết quả 
cách nhau một dấu cách

```python
n = float(input())
if  n >= 0 or n == int(n):
    floor_val = int(n)
else:
    floor_val = int(n) - 1

if n <= 0 or n == int(n):
    ceil_val = int(n)
else:
    ceil_val = int(n) + 1

if n >= 0:
    round_val = int(n+0.5) 
else:
    round_val = int(n-0.5)
    
print(f'{floor_val} {ceil_val} {round_val}')
```

Cấp số nhân  
Cho 4 số a, b, c, d. Hãy kiểm tra xem 4 số này có thể theo thứ tự tạo thành 1 
cấp số nhân với công bội nguyên theo đúng thứ tự a, b, c, d hay không?  
Gợi ý : Tìm công bội (b / a) nếu b chia hết cho a, sau đó lấy b nhân công bội 
và so sánh vs c, c nhân công bội và so sánh vs d.  

```python
a, b, c, d = map(int, input().split())
if b % a == 0:
    q = b // a
    if b * q == c and c * q == d:
        print('True')
    else:
        print('False')
else:
    print('False')
        
```

9. Chuyển đổi số thành chữ số: Nhập một số nguyên từ 0 đến 9, in ra tên chữ số 
tương ứng (0 -> "Không", 1 -> "Một", ...) – Xem trước match …case 

```python
n = int(input())
match n:
    case 0:
        chu = 'Không' 
    case 1:
        chu = 'Một' 
    case 2:
        chu = 'Hai' 
    case 3:
        chu = 'Ba' 
    case 4:
        chu = 'Bốn' 
    case 5:
        chu = 'Năm' 
    case 6:
        chu = 'Sáu' 
    case 7:
        chu = 'Bảy' 
    case 8:
        chu = 'Tám' 
    case 9:
        chu = 'Chín' 
    case _:
        chu = 'Không hợp lệ' 
print(f'Kết quả : {chu}')
```

Nhập vào chữ cái hoa A, in ra chữ cái thường liền trước chữ cái a thường tương 
ứng của A. Xét TH khi c = ‘A’ là đặc biệt.

```python
c = input()
if len(c) == 1 and 'A' <= c <= 'Z':
    chu_thuong = c.lower()
    match chu_thuong:
        case 'a':
            ket_qua = 'z'
        case _:
            ket_qua = chr(ord(chu_thuong) -1)
    print(f'Chữ cái thường trước {c} là : {ket_qua}')
else:
    print('Kông hợp lệ')
```
