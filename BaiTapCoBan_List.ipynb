### Bài 1: Tìm số lớn nhất của 1 list

```python
def max_of_list(lst):
    if lst:
        m = lst[0]
        for i in lst:
            if i > m:
                m = i
        return m
    return None
lst = list(map(int, input().split()))
print(max_of_list(lst))    
```

### Bài 7: Kiểm tra một list số nguyên có phải là dãy tăng, dãy giảm hay vô danh
1 3 5 6 7 8 9: Tang
9 8 7 6 5 4 0: Giam
1 2 3 7 1 0 2: Vo danh
1 2 2 4 5 7: Vo danh

```python
lst = list(map(int, input().split()))
if lst[0] > lst[1]:
    for i in range(len(lst)-1):
        if lst[i+1] >= lst[i]:
            print('vô danh')
            break
        else:
            print('tăng')
elif lst[0] < lst[1]:
    for i in range(len(lst)-1):
        if lst[i+1] <= lst[i]:
            print('vô danh')
            break
        else:
            print('giảm')
else:
    print('vô danh')
```
